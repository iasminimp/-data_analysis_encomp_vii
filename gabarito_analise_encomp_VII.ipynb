1. Importar a base de dados para o Python e visualizar o que ela possui

```python
# Trazer a base de dados para o python
import pandas as pd #pd apelido do pandas
tabela = pd.read_excel ("vendas.xlsx") #existem varios tipos (csv, html, sql)
display (tabela) #imprime a tabela de uma forma mais bonita
#verificar o que a tabela retorna pra gente
```

```python
#corrigir o nome da coluna: Valor Unitá¡rio
tabela.rename (columns={'Valor Unitá¡rio': 'Valor Unitário'}, inplace=True)
display (tabela) #print(tabela)
```



2. Panorama Geral


```python
#panorama geral
#faturamento total da empresa
faturamento_total = tabela["Valor Final"].sum()
print (faturamento_total)
```

3. Fazer uma analise TOP -> DOWN (verificar mais insights)

```python
#verificar o nome das colunas
print(tabela.columns)
```

```python
#fazer o faturamento por loja
faturamento_por_loja = tabela[["ID Loja","Valor Final"]]
print (faturamento_por_loja)
```

```python
#fazer um agrupamento de lojas somando o faturamento
#groupby = vai agrupar as lojas
#somando o  faturamento = sum()

faturamento_por_loja = tabela[["ID Loja","Valor Final"]].groupby("ID Loja").sum()
print (faturamento_por_loja)
```

```python
#verificar qual a loja que vende mais (vamos verificar qual produto esta saindo mais)
#colocar dentro do groupby uma lista
faturamento_por_produto = tabela[["ID Loja", "Produto", "Valor Final"]].groupby(["ID Loja", "Produto"]).sum()
display (faturamento_por_produto)
```

Solução: como temos que a loja que vendeu/ faturou mais foi a Loja A e o produto que ela mais vendeu foi a bermuda; Sendo assim a estratégia é aumentar o numero de bermudas nas outras lojas.
